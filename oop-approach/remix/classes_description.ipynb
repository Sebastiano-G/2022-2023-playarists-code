<code>class OCMetaProcessor</code>

<b>Attributes</b><br>
<code>batch_size</code> the size of the batches to process (int)<br>
<code>max_workers</code> the number of cpu workers to split the parallel processing<br>
<code>meta_path</code> the path to the OpenCitations Meta dump (needs to be unzipped)

<b>Methods</b><br>
<code>set_meta_path()</code> takes in input the <code>new_path</code> OpenCitations Meta dump and sets it as path<br>
<code>get_meta_path()</code> returns the path of the OpenCitations Meta dump dataset<br>
<code>set_batch_size()</code> takes in input an integer and changes the batch size<br>
<code>get_batch_size()</code> returns the current batch size<br>
<code>set_max_workers()</code> takes in input an integer and changes the number of workers<br>
<code>get_max_workers()</code> returns the number of workers
<code>process_files</code> takes in input an <code>ERIHPlusProcessor</code> object and a <code>DOAJProcessor</code> object and returns a DataFrame with OpenCitations Meta SSH Journals included in ERIH-PLUS dataset and their Open_Access information (OMID, EP_id, issn, Open_Access)
